```python
#Syntax of Lambda Function

#lambda arguments: expression

#arguments → Input variables (can be multiple, separated by commas)
#expression → Single line of code that is executed and returned

add = lambda x, y: x + y
print(add(5, 3))
```

```python

#using map fun

numbers = [1, 2, 3, 4, 5]
squared = list(map(lambda x: x**2, numbers))
print(squared)
```

map(function, iterable)
function → A function (can be a built-in function, lambda function, or user-defined function)
iterable → A sequence (list, tuple, etc.)

```python
numbers = ["1", "2", "3", "4"]
converted = list(map(int, numbers))  # Convert strings to integers
print(converted)
```

```python
list1 = [1, 2, 3]
list2 = [4, 5, 6]
summed = list(map(lambda x, y: x + y, list1, list2))
print(summed)
```

```python
words = ["hello","ddd"]
upper_words = list(map(str.upper, words))
print(upper_words)  # Output: ['HELLO', 'WORLD', 'PYTHON']
```

```python
def categorize_cholesterol(value):
    if value < 200:
        return "Normal"
    elif 200 <= value < 240:
        return "Borderline High"
    else:
        return "High"

csv["Cholesterol_Level"] = csv["Cholesterol"].apply(categorize_cholesterol)
print(csv)
```

```python
csv = pd.read_csv('Heart_Disease_.csv')
def new_cholesterol_category(value):
    if value < 180:
        return "Low"
    elif 180 <= value < 220:
        return "Normal"
    elif 220 <= value < 250:
        return "Borderline High"
    else:
        return "High"

# Update the same column with new values
csv["Cholesterol_Level"] = csv["Cholesterol"].apply(new_cholesterol_category)
print(csv)
```

```python
csv["Cardio_Index"] = (csv["Cholesterol"] * 0.6) + (csv["RestingBP"] * 0.4)
print(csv)
```


***Using df.groupby("column") in a Heart Disease Dataset***



The groupby() function in pandas groups data based on a column and allows us to compute statistics like mean, sum, count, min, max, etc.

```python
csv = pd.read_csv('Heart_Disease_.csv')
grouped = csv.groupby("StressLevel")
print(grouped)
```

```python
csv_grp=csv.groupby("SmokingStatus").count()
print(csv_grp)

```

```python
pivot_df = csv.pivot_table(index='Sex', columns='MaxHeartRate', values='BMI', aggfunc='mean')

print(pivot_df)
```
